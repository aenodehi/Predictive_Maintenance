#### Extracting **Utility** Tags from by_equipment.parquet

OK(12): FQI91501, FQI91502, FQI91503, FQI91504, FQI91505, FQI91506, FQI91507, FQI91508, FQI91509, FQI91510, FQI91512, FQI91513  
**NOT OK(3)**: FQI86502, FQI8503, FQI90025 

```python
search_term = 'FI91513'
results = []
```

```python
for file in parquet_files:
    df = pd.read_parquet(file)
    for col in df.select_dtypes(include='object').columns:
        mask = df[col].str.contains(search_term, na=False)
        if mask.any():
            matches = df[mask].copy()
            matches['source_file'] = file.name
            results.append(matches)
```

```python
if results:
    all_matches = pd.concat(results, ignore_index=True)
    print(f"Found {len(all_matches)} matching rows")
    display(all_matches)
else:
    print("No matches found")
```

```python
(all_matches['Value']>0).sum()
```

```python
total_rows = len(all_matches)
positive = (all_matches['Value'] > 0).sum()
print(f"Value>0: {positive}/{total_rows} ({100*positive/total_rows:.1f}%)")
```

```python
all_matches.to_parquet(f"../data/petrochem/{search_term}_u.parquet", index=False)
```

```python
parquet_dir_up = Path("/home/sagemaker-user/data/petrochem")
```

```python
u_files = list(parquet_dir_up.glob("*_u.parquet"))
p_files = list(parquet_dir_up.glob("*_p.parquet"))
```

```python
def create_summary(files):
    rows = []
    for f in files:
        df = pd.read_parquet(f)
        row = {
            'TAG':df['TAG'].iloc[0] if 'TAG' in df.columns else f.stem,
            'Unit':df['Unit'].iloc[0] if 'Unit' in df.columns else None,
            'Equipment':df['Equipment'].iloc[0] if 'Equipment' in df.columns else None,
            'Desc.':df['Desc.'].iloc[0] if 'Desc.' in df.columns else None,
            '#Value>0':(df['Value']>0).sum() if 'Value' in df.columns else 0
        }
        rows.append(row)
    return pd.DataFrame(rows)
```

```python
utility_table = create_summary(u_files)
product_table = create_summary(p_files)
```

```python
print(f"Found {len(u_files)} utility files")
print(f"Found {len(p_files)} product files")
```

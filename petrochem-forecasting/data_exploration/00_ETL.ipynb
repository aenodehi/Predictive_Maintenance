```python
from graphviz import Digraph
from IPython.display import display, HTML, IFrame

# Create a new directed graph
dot = Digraph(comment='ETL Pipeline', format='png')
dot.attr(rankdir='TB')  # Top to bottom direction
dot.attr('node', shape='box', style='rounded,filled', fillcolor='#E8D5F2', fontname='Arial')

# Define nodes
dot.node('etl', 'ETL Pipeline')
dot.node('logger', 'Logger')
dot.node('source', 'Data Source')
dot.node('processing', 'Data Processing')
dot.node('sink', 'Data Sink')
dot.node('console', 'Console')
dot.node('file', 'File')
dot.node('cloud', 'Cloud Service')
dot.node('database', 'Database')

# Define edges with labels
dot.edge('etl', 'logger', label='Logs Events')
dot.edge('etl', 'source', label='Extract')
dot.edge('etl', 'processing', label='Transform')
dot.edge('etl', 'sink', label='Load')

dot.edge('logger', 'console', label='Writes to')
dot.edge('logger', 'file', label='Writes to')
dot.edge('logger', 'cloud', label='Sends to')
#dot.edge('source', 'database', label='Sends to')
dot.edge('logger', 'database', label='Sends to')

# Render the graph
dot.render('etl_pipeline', cleanup=True)
print("Diagram saved as 'etl_pipeline.png'")

display(IFrame(src='etl_pipeline.png', width=1000, height=300))
```

```python
from pyvis.network import Network
import networkx as nx

# Create a NetworkX directed graph
G = nx.DiGraph()

# Add nodes with positions
nodes = {
    'ETL Pipeline': (0, 0),
    'Logger': (-300, 150),
    'Data Source': (-100, 150),
    'Data Processing': (100, 150),
    'Data Sink': (300, 150),
    'Console': (-400, 300),
    'File': (-250, 300),
    'Cloud Service': (-50, 300),
    'Database': (200, 300),
}

for node in nodes:
    G.add_node(node)

# Add edges with labels
edges = [
    ('ETL Pipeline', 'Logger', 'Logs Events'),
    ('ETL Pipeline', 'Data Source', 'Extract'),
    ('ETL Pipeline', 'Data Processing', 'Transform'),
    ('ETL Pipeline', 'Data Sink', 'Load'),
    ('Logger', 'Console', 'Writes to'),
    ('Logger', 'File', 'Writes to'),
    ('Logger', 'Cloud Service', 'Sends to'),
    ('Logger', 'Database', 'Sends to'),
]

for source, target, label in edges:
    G.add_edge(source, target, label=label)

# Create Pyvis network
net = Network(
    directed=True,
    height='400px',
    width='100%',
    notebook=False
)

net.from_nx(G)

# Set positions
for node in net.nodes:
    label = node['label']
    if label in nodes:
        x, y = nodes[label]
        node['x'] = x * 1.1
        node['y'] = y - 170 
        node['physics'] = False

# Customize nodes
for node in net.nodes:
    node['color'] = '#E8D5F2'
    node['font'] = {'size': 14}
    node['borderWidth'] = 2
    node['shape'] = 'box'
    node['size'] = 10
    if node['label'] == 'ETL Pipeline':
        node['color'] = '#9966CC'
        node['font']['size'] = 14
        node['size'] = 20

smooth_edges = [
    ('ETL Pipeline', 'Logger'),
    ('Logger', 'Console'),
]

# Customize edges
for edge in net.edges:
    source, target = edge['from'], edge['to']
    edge['arrows'] = {'to': {'enabled': True, 'scaleFactor': 0.5}}
    edge['font'] = {'size': 11}
    #edge['smooth'] = {'type': 'continuous'}
    #edge['smooth'] = False
    edge['color'] = '#666666'
    if (source, target) in smooth_edges:
        edge['smooth'] = {'type': 'continuous'}
    else:
        edge['smooth'] = False

# Physics simulation disabled for fixed layout
net.physics = False

# Save as HTML
output_file = 'etl_pipeline_interactive.html'
net.write_html(output_file)
print(f"Interactive diagram saved as '{output_file}'")
display(IFrame(src='etl_pipeline_interactive.html', width=1000, height=500))
```
